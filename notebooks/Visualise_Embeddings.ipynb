<h1 align="center">Visualising Embeddings</h1>

In this notebook we compare the different embeddings by looking at their correlation heatmaps. We can observe that despite being produced via very different approaches all of the embeddings exhibit similar correlation structures.

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib notebook

def plot_heatmap(data):
    corr = data.corr()
    plt.figure(figsize=(8,8))
    g = sns.heatmap(corr, 
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values, 
                cbar=False)
    for tick in g.xaxis.get_major_ticks()[1::2]:
        tick.set_pad(25)
    for tick in g.yaxis.get_major_ticks()[1::2]:
        tick.set_pad(25)
    plt.show()
```

```python
matscholar = pd.read_json('matscholar_embedding.json')

plot_heatmap(matscholar)
```

```python
megnet = pd.read_json('megnet_embedding.json')

plot_heatmap(megnet)
```

```python
cgcnn = pd.read_json('cgcnn_embedding.json')

plot_heatmap(cgcnn)
```

```python
elem = pd.read_json('elem_embedding.json')

plot_heatmap(elem)
```

```python
elem_low = pd.read_json('elem_low_embedding.json')

plot_heatmap(elem_low)
```
