# Visualising Embeddings

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
matscholar = pd.read_json('matscholar_embedding.json')
megnet = pd.read_json('megnet_embedding.json')
cgcnn = pd.read_json('cgcnn_embedding.json')
elem = pd.read_json('elem_embedding.json')
elem_low = pd.read_json('elem_low_embedding.json')
```

```python
corr = matscholar.corr()
plt.figure()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

```python
plt.figure()
corr = megnet.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

```python
plt.figure()
corr = cgcnn.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

```python
plt.figure()
corr = elem.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

```python
plt.figure()
corr = elem_low.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```
